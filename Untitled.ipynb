```python
import re
Bathrooms=[]
Facing=[]
h = b.find_all('ul',attrs={'class':'listing-details'})
for i in h:
    i= i.text
    Bathrooms_count = re.findall(r'(\d) Bathrooms',i)
    #print(Bathrooms_count)
    if Bathrooms_count:
        Bathrooms.append(int(Bathrooms_count[0]))
    else:
        Bathrooms.append(0)
    facing_count = re.findall(r'(North|South|East|West|NorthEast|SouthEast|SouthWest)facing',i)
    if facing_count:
        Facing.append('facing_count{0}')
    else:
        Facing.append('None')
print(Bathrooms)
print(Facing)
print(len(Bathrooms))
print(len(Facing))        
```
